```python
#Hexbin Plot(using jointplot)


X, y = make_blobs(n_samples=4000, centers=3, n_features=2, random_state=0)

blobs = pd.DataFrame(dict(zip(("x", "y", "group"), (X[:,0], X[:,1], y))))

g = sns.jointplot(blobs.x, blobs.y, kind="hex")
```

```python
#Scatterplot with reduced size

sns.scatterplot(x=blobs.x, y=blobs.y, size=10)
```

```python
#Jointplot with kernal density estimation

g = sns.jointplot(blobs.x, blobs.y, kind="kde", height=7, space=0)
```

```python
#Code to create a Search Function

def searchy(pattern, string, ignore_case):
    """Find index of pattern match in string. Returns empty list if not found."""
    if ignore_case:
        pattern = pattern.lower()
        string = string.lower()

    for i in range(len(string)):
        for j in range(len(pattern)):
            if string[i+j] != pattern[j]:
                return print( '%s is a substring of %s' %(pattern, string))
                break
            elif j == len(pattern) - 1:
                return ['']
    
```

```python
#When we are not ignoring the case
rain_in_spain = 'The rain\n in spain \nfalls mainly on the plane'

searchy('rain', rain_in_spain, ignore_case=True)
```

```python
#When we are ignoring the case
rain_in_spain = 'The rain\n in spain \nfalls mainly on the plane'

searchy('Rain', rain_in_spain, ignore_case=False)
```
